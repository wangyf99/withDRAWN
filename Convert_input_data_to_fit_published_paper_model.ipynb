```python
import pandas as pd
```

```python
train1 = pd.read_csv("train1.csv")
test1 = pd.read_csv("test1.csv")
val = pd.read_csv("val.csv")
```

```python
drug_labels_smiles_final = pd.read_csv("tox_labels_smiles_final.csv")
drug_labels_smiles_final
```

```python
drug_labels_smiles_final = drug_labels_smiles_final.loc[:, ['drug', 'withdrawn', 'smiles']]
```

```python
drug_labels_smiles_final.columns = ['drug', 'withdrawn_class', 'smiles']
```

```python
drug_labels_smiles_final
```

```python
train1.columns = ['drug']
```

```python
merged_train = train1.merge(drug_labels_smiles_final, on='drug', how='inner')
```

```python
test1.columns = ['drug']
merged_test = test1.merge(drug_labels_smiles_final, on='drug')
merged_test
```

```python
val.columns = ['drug']
merged_val = val.merge(drug_labels_smiles_final, on='drug')
merged_val
```

```python
merged_train.to_csv("merged_train.csv")
merged_test.to_csv("merged_test.csv")
merged_val.to_csv("merged_val.csv")
```
