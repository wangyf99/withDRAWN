```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
```

```python
def get_smiles_from_chembl(chembl_id):
    try:
        molecule = new_client.molecule.filter(chembl_id=chembl_id).only("molecule_structures")
        if molecule:
            return molecule[0]['molecule_structures']['canonical_smiles']
        else:
            return None
    except Exception as e:
        print(f"Error for {chembl_id}: {e}")
        return None
```

```python
df = pd.read_csv("tox_labels.csv")
df
```

```python
chembl_df = df[df['drug'].str.startswith('chembl')]
```

```python
chembl_df
```

```python
smiles_df = chembl_df['drug'].apply(get_smiles_from_chembl)
```

```python
smiles_df
```

```python
chembl_df.loc[:, 'smiles'] = smiles_df
```

```python
chembl_df
```

```python
chembl_df.to_csv("tox_labels_chembl.csv")
```
