```python
from chembl_webresource_client.new_client import new_client

molecule = new_client.molecule
res = molecule.filter(pref_name__iexact=['MIDODRINE HYDROCHLORIDE '])#.only(['molecule_structures'])
chembl_id = next(iter(res))#['molecule_structures']['canonical_smiles']
print(chembl_id)
# for i in res:
#     print(res)
```

```python
def get_chembl_from_name(row):
    try:
        molecule = new_client.molecule
        res = molecule.filter(pref_name__iexact=row['drug'])
        output = next(iter(res))
        if output is not None:
            molecule_structure = output['molecule_structures']
            if molecule_structure is not None and molecule_structure['canonical_smiles'] is not None:
                return molecule_structure['canonical_smiles']
            else:
                return None
        else:
            return None
        
    except:
        print(f"drug {row['drug']} didn't get any chembl ID")
        return None
        
```

```python
df = pd.read_csv("tox_labels.csv")
df
```

```python
df_name = df[~df['drug'].str.startswith("chembl")].copy()
```

```python
df_name
```

```python
df_name['smiles'] = df_name.apply(get_chembl_from_name, axis=1)
df_name
```

```python
df_name.to_csv('tox_labels_names_smiles.csv')
```
